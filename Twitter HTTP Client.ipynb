```python
# import packages
import tweepy # route Live tweets to a TCP socket server
from tweepy import OAuthHandler # Authentication protocol of Twitter API
from tweepy import Stream
from tweepy.streaming import StreamListener # used for listening to streaming data getting from socket 
import socket # create socket connection with local server
import json # twitter data would come from json format 
```

```python
# Enter Twitter API keys and tokens
# Accessed from http://apps.twitter.com/
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''
```

```python
# create a class called TweetsListener to listen to stream
class TweetsListener(StreamListener):
    def __init__(self, csocket):
        self.client_socket = csocket
    def on_data(self, data):
        try:
            msg = json.loads( data ) # read the data that come from socket  
            self.client_socket.send( msg['text'].encode('utf-8') ) # only use the 'text' column of msg and use 'utf-8' to encode message (text include the content of each tweets)
            # send the content to a socket
            print(msg['text'].encode('utf-8'))
            return True
        except BaseException as e:
            print("Error on_data: %s" % str(e)) # handle exception
        return True
    def on_error(self, status): 
        print(status)
        return True
```

```python
def sendData(c_socket):
    auth = OAuthHandler(consumer_key, consumer_secret) # doing authentication check using OAuthHandler.
    auth.set_access_token(access_token,access_secret) # authentication key
    twitter_stream = Stream(auth, TweetsListener(c_socket)) # to get the twitter stream data (live)
    twitter_stream.filter(track = ['football']) # filter the keyword
```

```python
# this functions checks if a port is being connected or not
def isOpen(ip,port):
   s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
   try:
      s.connect((ip, int(port)))
      s.shutdown(2)
      return True
   except:
      return False

isOpen("localhost",9008)
```

```python
# Socket Creation
s = socket.socket() # create a socket object
host = "localhost" # local machine name
port = 9008 # reserve a port for service (can be anything that without being connected, unique for this service)
s.bind((host,port)) # bind the host and port address to the socket
print("Listening on port: %s" % str(port))
```

```python
s.listen(5) # socket will wait forclient connection
```

```python
c, addr = s.accept() # Establish a connection 
print("Received request from: " + str(addr))
```

```python
sendData(c) # will send the Tweets being filtered by keyword to socket 'c' in real time
```
